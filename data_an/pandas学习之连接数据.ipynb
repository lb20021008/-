```python
#合并方式二: merage() 基于键的横向连接(增加列)
df1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                        'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
#
df2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                        'key2': ['K0', 'K0', 'K0', 'K0'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
# 基于键的链接,两个df对象根据key1链接(df1.key1==df2.key1),连接后只会有一个key1,inner方式只保留匹配数据
pd.merge(df1,df2,on='key1',how='inner')
```

```python
# 两个df对象根据key1和key2链接(df1.key1==df2.key1 and df1.key2==df2.key2),连接后只会有一个key1,key2,inner方式只保留匹配数据
pd.merge(df1,df2,on=['key1','key2'],how='inner')
```

```python
# 两个df对象根据key1和key2链接(df1.key1==df2.key1 and df1.key2==df2.key2),连接后只会有一个key1,key2,outer方式保留匹配数据和源数据
pd.merge(df1,df2,on=['key1','key2'],how='outer')
```

```python
# 两个df对象根据key1和key2链接(df1.key1==df2.key1 and df1.key2==df2.key2),连接后只会有一个key1,key2,left方式只保留左表所有数据和右表匹配数据
pd.merge(df1,df2,on=['key1','key2'],how='left')
```

```python
# 两个df对象根据key1和key2链接(df1.key1==df2.key1 and df1.key2==df2.key2),连接后只会有一个key1,key2,left方式只保留右表所有数据和左表匹配数据
pd.merge(df1,df2,on=['key1','key2'],how='right')
```
