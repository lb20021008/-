```python
import numpy as np
```

```python
# 传入列表创建ndarry
list_a = [11,22,33]
arr1 = np.array(list_a)
print(arr1)
print(type(arr1))
```

```python
# 用arange方法创建数组
arr2 = np.arange(0,10,2,dtype=np.int64)
print(arr2)
print(type(arr2))  # <class 'numpy.ndarray'>
```

```python
# 用random.rand方法随机生成多维数组
arr3 = np.random.rand(3,4) # 生成一个三行四列,范围在[0,1)之间的矩阵
print(arr3)
print(type(arr3))
```

```python
# 用random.randint生成多维整形数组
arr4 = np.random.randint(0,10,(3,4))# 生成一个范围是[0,10) 三行四列的数组
print(arr4)
print(type(arr4))
```

```python
# 用random.uniform生成多维浮点数数组
arr5 = np.random.uniform(0,10,(3,4))# 范围[0,10)形状3行4列
print(arr5)
print(type(arr5))
```

```python
# 试一试random.randn
arr6 = np.random.randn(3,4)
print(arr6)
```

```python
# 改变数组类型
arr7 = np.arange(0,10,dtype=np.int64)
print(arr7)
print(arr7.dtype)
arr7 = arr7.astype(np.float64) # astype方法改变dtype元素类型为float64
print(arr7)
print(arr7.dtype)
```

```python
# 等比和等差数列
arr8 = np.logspace(0,3,4) # 按照10的指数从0开始(10的0次幂)到3(10的三次幂)结束
print(arr8)
print(type(arr8))

arr9 = np.linspace(0,10,10)
print(arr9)
print(type(arr9))
```

```python
# numpy内置函数
arr10 = np.random.randn(3,5)
print(arr10)
print(np.ceil(arr10))
print(np.floor(arr10))
print(np.abs(arr10))
print(np.rint(arr10))
print(np.multiply(arr10,arr10))
print(np.divide(arr10,arr10))
print(np.where(arr10>1,1,0))
```

```python
# np统计函数
arr1 = np.arange(1,13,dtype=np.int64).reshape(3,4)
print(arr1)
print(np.sum(arr1))
print(np.sum(arr1,axis=0))
print(np.sum(arr1,axis=1))
print(np.argmax(arr1))
print(np.argmin(arr1))
print(np.cumsum(arr1))
print(np.cumprod(arr1))
```

```python
# 数组去重和排序
arr2 = np.array([[1,3,2],[4,9,3]])
arr2_un = np.unique(arr2)
print(arr2_un)

arr2 = np.array([[1,3,2,7,5,9,1]])
arr2_sort = np.sort(arr2)
print(arr2_sort)
```

```python
# 矩阵相乘(行列数不一致)
arr1 = np.array([[1,2,3],[4,5,6]])
arr2 = np.array([[1,2],
                 [3,4],
                 [5,6]])
# 方法一
print(arr1@arr2)
# 方法二
print(np.dot(arr1,arr2))
print(arr1.dot(arr2))
```
