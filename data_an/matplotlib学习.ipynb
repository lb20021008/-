```python
import random

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.sans-serif'] = ['SimHei']    # 如果是Mac本, 不支持SimHei的时候, 可以修改为 'Microsoft YaHei' 或者 'Arial Unicode MS'
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(8,4),dpi=80)
plt.plot([i for i in range(1,8)],[11,25,12,32,15,16,9])
plt.show()
```

```python
plt.figure(figsize=(20,10),dpi=200)
# 准备数据
x=range(60)
y = [random.uniform(15,18) for i in x]
#绘制图表用折线方式(plot)
plt.plot(x,y,color='red')
# 设置坐标限度
x_ticks = [f'11点{i}分' for i in x]
y_ticks=range(40)


# 设置刻度间隔为5
plt.xticks(x[::5],x_ticks[::5])
plt.yticks(y_ticks[::5])
# 绘制网格
plt.grid(linestyle='--',alpha=0.3)
# 添加xy轴标签
plt.xlabel('时间',fontsize=10)
plt.ylabel('温度',fontsize=10)
# 添加标题
plt.title('温度曲线')
# 保存图
plt.savefig('./my_pic')
# 画图
plt.show()
```

```python
# 两个折线图制在一个坐标系里
plt.figure(figsize=(20,10),dpi=200)

x = range(60)
y = [random.uniform(15,18) for i in x]
y_2 = [random.uniform(-7,3) for i in x]

x_s_l = [f'11点{i}分' for i in x]
y_s_l = range(-10,40)

plt.plot(x,y,color='blue',label = '上海',linestyle='-')
plt.plot(x,y_2,color='red',label = '北京',linestyle = '--')

plt.xticks(x[::5],x_s_l[::5])
plt.yticks(y_s_l[::5])
plt.grid(linestyle='--',alpha=0.3)

# 给两根线取名作图例
plt.xlabel('时间')
plt.ylabel('温度')

plt.title("北京上海一小时内温度变化折线图")
# 设置图里的为值,默认为最佳位置
plt.legend()

plt.show()
```

```python
# 将折线图花在两个坐标系里
fig,t = plt.subplots(nrows=1,ncols=2,figsize=(60,30),dpi=200)

# 造值
x = [i for i in range(60)]

y_s = [random.uniform(15,18) for i in x]
y_b = [random.uniform(-7,3) for i in x]

t[0].plot(x,y_s,color='blue',label = '上海',linestyle='-')
t[1].plot(x,y_b,color='red',label = '北京',linestyle = '--')

x_t = [f'11点{i}分' for i in x]
y_t_s = range(15,19)
y_t_b = range(-10,10)

# 注意方法
t[0].set_xticks(x[::5],x_t[::5])
t[1].set_xticks(x[::5],x_t[::5])

t[0].set_yticks(y_t_s[::3])
t[1].set_yticks(y_t_b[::3])

t[0].grid(linestyle='--',alpha=0.3)
t[1].grid(linestyle='--',alpha=0.3)

#注意方法
t[0].set_title('1小时温度')
t[0].set_xlabel('时间')
t[0].set_ylabel('温度')

t[1].set_title('1小时温度')
t[1].set_xlabel('时间')
t[1].set_ylabel('温度')

t[0].legend()
t[1].legend()

plt.savefig('./my_pic2')

plt.show()
```
