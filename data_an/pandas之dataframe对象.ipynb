```python
import pandas as pd
import numpy as np
from xarray.util.generate_ops import inplace
```

```python
# 字典加列表
dict1 = {
    'name':['lisi','zhangsan','wangwu'],
    'age':[11,20,30]
}
pd.DataFrame(data=dict1)
```

```python
# 元组加列表
t1 = (
    ['lisi',12],
    ['zhangsan',30],
    ['wangwu',20]
)
pd.DataFrame(data=t1,columns=['name','age'])
```

```python
# numpy创造df对象
arr1 = np.arange(12).reshape(3,4)
pd.DataFrame(data=arr1)
```

```python
# 创建一个datafame
d1 = pd.DataFrame(np.random.randint(40,101,(10,5)))
cul_n = ['语文','数学','英语','物理','化学']
in_n = ['同学'+str(i) for i in range(d1.shape[0])]
# d1.index = in_n
# d1.columns = cul_n
# print(d1)
```

```python
# 设置行列索引
d1.rename(
    columns={i:cul_n[i] for i in range(d1.shape[1])},
    index={i:in_n[i] for i in range(d1.shape[0])},
    inplace = True
)
print(d1)
```

```python
# 重置索引时保留原索引列
d1.reset_index(drop=False)
# 充值索引时不保留元索引列
d1.reset_index(drop=True)
```

```python
# 将语文和英语设为索引
d1.set_index(['语文','英语'])
```
