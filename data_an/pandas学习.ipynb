```python
# 相对路径
df = pd.read_csv(r'./data/1960-2019全球GDP数据.csv', encoding='gbk')
print(df)
# 绝对路径
# df_s = pd.read_csv(r'C:\Users\27700\PycharmProjects\data_an\data\1960-2019全球GDP数据.csv', encoding='gbk')
```

```python
china_df = df[df.country=='中国'].set_index('year') # 筛选出国家为中国的数据
```

```python
usa_df = df[df.country == '美国'].set_index('year')
```

```python
jp_df = df[df.country == '日本'].set_index('year')
```

```python
china_df.GDP.plot() # 根据year和GDP画图
usa_df.GDP.plot()
jp_df.GDP.plot()
```

```python
china_df.rename(columns={'GDP':'中国'},inplace=True)
usa_df.rename(columns={'GDP':'美国'},inplace=True)
jp_df.rename(columns={'GDP':'日本'},inplace=True)
```

```python
china_df.中国.plot(legend=True)
usa_df.美国.plot(legend=True)
jp_df.日本.plot(legend=True)
```
