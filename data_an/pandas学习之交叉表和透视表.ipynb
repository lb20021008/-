交叉表演示

```python
import pandas as pd
import numpy as  np
# 创建一个示例数据集
data = {
    '性别': ['男', '女', '男', '女', '男', '女', '女', '男'],
    '购买': ['是', '否', '是', '是', '否', '否', '是', '否']
}

df = pd.DataFrame(data)
df
```

```python
# 将性列名中的值作为行索引,购买的值作为列索引,统计其出现次数,[女,否]出现两次,[女,是]出现两次......
pd.crosstab(index=df['性别'],columns=df['购买'])
```

透视表演示

```python
# 数据准备
import pandas as pd

data = {
    '性别': ['男', '女', '男', '女', '男', '女'],
    '购买': ['是', '否', '是', '是', '否', '否'],
    '金额': [100, 150, 200, 130, 160, 120]
}
df = pd.DataFrame(data)
df
```

```python
# 创建透视表
# 将性列名中的值作为行索引,购买列中的值作为列索引,统计各个值分类下的金额(values=金额),返回平均值或者最大值,最小值总和......默认统计平均值aggfunc=mean
pd.pivot_table(df,values='金额',index='性别',columns='购买')
```
