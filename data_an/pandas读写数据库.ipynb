```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('./data/csv示例文件.csv',encoding='gbk',index_col=0)
# 获取引擎对象
engine = create_engine('mysql+pymysql://root:admin@localhost:3306/db_8?charset=utf8')
# 将数据导入数据库
df.to_sql('my_table',engine,if_exists='append',index=False)
```

```python
# 将数据库的整张表写入dataframe对象
sql_df=pd.read_sql('my_table',engine)
sql_df
```

```python
# 通过查询语句写入dataframe对象
sql_df = pd.read_sql('select AKA from my_table',engine)
sql_df
```
