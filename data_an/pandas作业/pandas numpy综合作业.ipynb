```python
# 导包
import pandas as pd
import numpy as np
```

使用 `np.arange()` 创建一个从 0 到 9（包含 0，不包含 10）的一维数组。

```python
np.arange(0,10)
```

使用 `np.arange()` 创建一个从 5 到 20（步长为 3）的数组。

```python
np.arange(5,20,3)
```

创建一个 3×4 的二维数组，并打印其 `shape`。

```python
np.arange(12).reshape(3,4)
```

给定数组 `a = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])`，打印它的 `shape`。

```python
a = np.array([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(a.shape)
```

创建一个整数数组 `[1, 2, 3]`，并打印其数据类型

```python
print(np.arange(1,4).dtype)
```

创建一个浮点数组，并显式指定 `dtype` 为 `np.float32`，然后打印其 `dtype`。

```python
print(np.arange(1,4,dtype=np.float32).dtype)
```

创建一个标量（0 维）数组 `np.array(5)`，并打印其维度数。

```python
print(np.array(5).ndim)
```

创建一个三维数组 `np.ones((2, 3, 4))`，并打印其 `ndim`。

```python
print(np.ones((2,3,4)).ndim)
```

给定数组 `a = np.array([3, 1, 4, 1, 5, 9, 2])`，找出最大值的索引。

```python
a = np.array([3, 1, 4, 1, 5, 9, 2])
np.argmax(a)
```

给定二维数组 `b = np.array([[1, 2, 3], [6, 5, 4]])`，找出每列最大值的行索引

```python
b = np.array([[1, 2, 3], [6, 5, 4]])
for i in range(b.shape[0]):
    print(f'第{i+1}行最大值索引为{np.argmax(b[i])}')
```

对数组 `c = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5])` 去重并排序（使用 `np.unique`）。

```python
c = np.array([3, 1, 4, 1, 5, 9, 2, 6, 5])
print(np.unique(c))
print(c)
```

对数组 `d = np.array([9, 3, 7, 1])` 进行升序排序（使用 `np.sort`），并保留原数组不变。

```python
d = np.array([9, 3, 7, 1])
print(np.sort(d))
print(d)
```

创建一个 2×3 的矩阵 A 和一个 3×2 的矩阵 B，计算 `A @ B`

```python
a = np.random.randint(0,6,(2,3))
b = np.random.randint(0,6,(3,2))
print(a@b)
```

尝试用 `@` 计算两个不能相乘的矩阵（如 2×3 和 2×2），观察报错信息。

```python
a = np.random.randint(0,6,(2,3))
b = np.random.randint(0,6,(2,2))
print(a@b)
```

创建一个包含三列数值数据的 DataFrame（例如：`'A': [1, 2, 3, 4]`, `'B': [10, 20, 30, 40]`, `'C': [100, 200, 300, 400]`），并调用 `.describe()` 查看统计摘要。

```python
data_d = {'A': [1, 2, 3, 4], 'B': [10, 20, 30, 40], 'C': [100, 200, 300, 400]}
df = pd.DataFrame(data_d)
print(df)
df.describe()
```

创建一个包含缺失值的 DataFrame（例如：`'X': [1, 2, None]`, `'Y': ['a', None, 'c']`），调用 `.info()` 查看各列的数据类型和非空计数。

```python
df2 = pd.DataFrame({'X': [1, 2, None], 'Y': ['a', None, 'c']})
df2.info
```
