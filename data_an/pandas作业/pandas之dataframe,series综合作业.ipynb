假设我们有一个电影数据集`movies.csv`，包含以下列：`movie_id`（电影ID）、`title`（电影标题）、`genre`（电影类型）、`release_year`（发行年份）、`rating`（评分）、`duration`（时长，分钟）。

**题目**：导入Pandas库。

```python
import pandas as pd
from sqlalchemy.dialects.mssql.information_schema import columns
```

**题目**：读取`movies.csv`文件并存储为DataFrame对象。
