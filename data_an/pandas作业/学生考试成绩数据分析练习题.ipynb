 背景：某班级 5 名学生的 3 门科目（数学、英语、物理）期中考试成绩如下，使用 Numpy 进行数据处理与分析。

```python
# 数据准备
import numpy as np

# 创建成绩数据（单位：分）
scores_data = {
    '学生ID': ['S001', 'S002', 'S003', 'S004', 'S005'],
    '数学': [85, 92, 76, 88, 95],
    '英语': [78, 85, 82, 90, 88],
    '物理': [90, 87, 91, 79, 83]
}

# 转换为ndarray（以学生ID为索引）
scores_ndarray = np.array([
    [85, 78, 90],
    [92, 85, 87],
    [76, 82, 91],
    [88, 90, 79],
    [95, 88, 83]
], dtype=np.float64)

# 设置学生ID为行索引（实际操作中可通过花式索引实现）
student_ids = np.array(['S001', 'S002', 'S003', 'S004', 'S005'])
subjects = np.array(['数学', '英语', '物理'])
```

1. 查看成绩数组的维度、形状、元素类型及总个数

```python
print(scores_ndarray.ndim) #维度
print(scores_ndarray.shape) # 形状
print(scores_ndarray.dtype) # 元素类型
print(scores_ndarray.size) # 个数
```

2. 将成绩数组转换为整数类型，并查看每个元素的字节大小

```python
scores_ndarray.dtype = 'int64'
print(scores_ndarray.dtype)
print(scores_ndarray.itemsize)
```

3. 计算所有学生的数学平均分、物理总分及英语最高分

```python
print(np.average(np.array(scores_data['数学'])))
print(np.max(np.array(scores_data['物理'])))
print(np.max(np.array(scores_data['英语'])))
```

对物理成绩进行排序，并找到最低分的索引

```python
p = np.sort(np.array(scores_data['物理']))
print(np.argmin(p))
```

使用where函数将 90 分以上的成绩标记为 'A'，其余为 'B'

```python
np.where(scores_ndarray>90,'A','B')
```

计算每个学生的总分（行方向求和）与每科的平均分（列方向求平均）

```python
print(scores_ndarray.sum(1))
print(np.average(scores_ndarray,0))
```

使用cumsum计算每个学生的累计得分？

```python
sum_n = np.cumsum(scores_ndarray,1)
print([i[2] for i in sum_n])
```
