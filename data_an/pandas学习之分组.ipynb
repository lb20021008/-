```python
import pandas as pd
import numpy as np
```

```python
# 导入数据
df = pd.read_csv('./data/uniqlo.csv')
df
```

```python
# 按'city','channel'分组直接打印是地址
df.groupby(['city','channel'])
```

```python
# 按'city','channel'分组,计算出这个分组下每个组最大营销额(revenue)
df.groupby(['city','channel']).agg({
    'revenue':'max'
})
```

```python
# 按'city','channel'分组,计算出这个分组下每个组总营销额(revenue)
df.groupby(['city','channel'])['revenue'].sum()
```

```python
# 写法二
df.groupby(['city','channel'])[['revenue']].sum()
```

```python
# 按'city','channel'分组,计算出这个分组下每个组总营销额(revenue)和总成本(unit_cost)
df.groupby(['city','channel']).agg({
    'revenue':'sum',
    'unit_cost':'sum'
})
```

```python
# 写法二
df.groupby(['city','channel'])[['revenue','unit_cost']].sum()
```

```python
df.groupby(['city','channel']).agg({
    'revenue':'sum',
    'unit_cost':'mean'
})
```

```python
# 按城市分组过滤出营销额平均值大于200的
df.groupby(['city']).filter(lambda x:x['revenue'].mean()>200)
```
