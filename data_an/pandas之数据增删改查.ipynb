```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
```

```python
df = pd.read_csv('./data/1960-2019全球GDP数据.csv',encoding='gbk')
df
```

```python
# 复制df前五行作为一个新df对象
n_df = df.head().copy()
```

```python
# 添加一个c1列,所有值为32
n_df['c1'] = 32
# 添加一个c2列,值为1,2,3,4,5
n_df['c2'] = [1,2,3,4,5]
# 添加一个新列用lambda表达式赋值
n_df['c3'] = (lambda :2500)()
# 添加一个列,用f1函数赋值
def f1():
    return 2000
n_df['c4'] = f1()
n_df
```

```python
# 删除行索引为0和1的行
n_df.drop(index=[0,1],inplace=True)
n_df
```

```python
# del方法删除c1列
del n_df['c1']
n_df
```

```python
# drop方法删除多个列
n_df.drop(columns=['c2','c3','c4'],inplace=True)
```

```python
n_df
```

```python
# concat方法拼接将多个n_df拼接
df2 = pd.concat([n_df,n_df,n_df,n_df,n_df])
df2
```

```python
# 去重
df2.drop_duplicates()
```

```python
# 删除包含country列中的重复值的行
df2.drop_duplicates('country')
```
