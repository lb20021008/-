```python
import pandas作业 as pd
import numpy as np
```

```python
# 列表方式创建Series对象,设置索引为abcd
pd.Series([1,2,3,4],index=['a','b','c','d'])
```

```python
# 元组方式创建Series对象,设置索引为abcd
pd.Series((1,2,3,4),index=['a','b','c','d'])
```

```python
# 字典方式创建Series对象,设置索引为abc
pd.Series({'a':1,'b':2,'c':3})
```

```python
# ndarray方式创建Series对象,这里弄了一个等比数列
pd.Series(np.logspace(0,4,5))
```

```python
# 0-26对应abcd...xyz
s1 = pd.Series([i for i in range(26)],index=[chr(i) for i in range(97,123)])
print(s1.index)
print(s1.values)
# 修改索引d的值为13
s1['d'] = 13
print(s1['d'])
print(s1)
```
